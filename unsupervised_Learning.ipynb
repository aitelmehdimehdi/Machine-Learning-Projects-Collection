**clustering**

```python
def best_param_KMEANS(X, min_clusters=2 , max_clusters=10):
    max_power = -1
    min_n_clusters = 1

    for n in range(min_clusters, max_clusters + 1):
        kmeans_model = sk.cluster.KMeans(n_clusters=n)
        kmeans_model.fit(X)
        labels = kmeans_model.labels_
        sc = sk.metrics.silhouette_score(X, labels)
        if sc > max_power:
            max_power = sc
            min_n_clusters = n
    print(f'Performance de meilleure modÃ¨le : [ silhouette_coefficient = {max_power}')
    return min_n_clusters
```

```python
def best_param_DBSCAN(X,min_eps=0.01 , max_eps=1 , min_min_samples=2 , max_min_samples=3):

  eps_interval = np.linspace(min_eps,max_eps,num=10).astype('float')
  min_samples_interval = np.unique(np.linspace(min_min_samples,max_min_samples,num=5).astype('int'))

  min_eps = 0.001
  min_samples = 0
  silhouette_performance = -1
  davies_bouldin_score_performance = 1000
  calinski_harabasz_score_performance = 0

  for eps in eps_interval:
    for min_samples in min_samples_interval:
      temp_model = sk.cluster.DBSCAN(eps=eps, min_samples=min_samples)
      temp_model.fit(X)
      labels = temp_model.labels_

      # Check if there are at least 2 unique labels (excluding -1 for noise)
      n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
      #print(f'{eps=}, {min_samples=}, {n_clusters=}')  # Add this line for debugging

      if n_clusters >= 2:
        param1 = sk.metrics.silhouette_score(X, labels)
        param2 = sk.metrics.davies_bouldin_score(X, labels)
        param3 = sk.metrics.calinski_harabasz_score(X, labels)
        if param1 > silhouette_performance:
          if param2 < davies_bouldin_score_performance:
            if param3 > calinski_harabasz_score_performance:
              silhouette_performance = param1
              davies_bouldin_score_performance = param2
              calinski_harabasz_score_performance = param3
              min_eps = eps
              min_samples = min_samples
  print(f'Performance de meilleure modÃ¨le : [ silhouette_coefficient = {silhouette_performance} ; davies_bouldin_score_coefficient = {davies_bouldin_score_performance} ; calinski_harabasz_score_coefficient = {calinski_harabasz_score_performance} ]')

  return min_eps , min_samples
```

```python
kmeans_nclusters_PCA = best_param_KMEANS(xByPCA)
esp_dbscan_PCA , samples_dbscan_PCA = best_param_DBSCAN(xByPCA,min_eps=0.5,min_min_samples=10)

kmeans_nclusters_TSNE = best_param_KMEANS(xByTSNE,10,15)
esp_dbscan_TSNE , samples_dbscan_TSNE = best_param_DBSCAN(xByTSNE)
```

```python
esp_dbscan_PCA
```

```python
KMeans_model_pca = sk.cluster.KMeans(n_clusters=kmeans_nclusters_PCA)
DBSCAN_model_pca = sk.cluster.DBSCAN(eps=esp_dbscan_PCA, min_samples=samples_dbscan_PCA)

KMeans_model_tsne = sk.cluster.KMeans(n_clusters=kmeans_nclusters_PCA)
DBSCAN_model_tsne = sk.cluster.DBSCAN(eps=0.5, min_samples=5)
```

```python
KMeans_model_pca.fit(xByPCA)
DBSCAN_model_pca.fit(xByPCA)

KMeans_model_tsne.fit(xByTSNE)
DBSCAN_model_tsne.fit(xByTSNE)
```

```python
X_kmeans_pca = KMeans_model_pca.labels_
X_dbscan_pca = DBSCAN_model_pca.labels_

X_kmeans_tsne = KMeans_model_tsne.labels_
X_dbscan_tsne = DBSCAN_model_tsne.labels_
```

```python
pca_df = pd.DataFrame(xByPCA , columns=['x','y'])
pca_df['kmeans_labels'] = X_kmeans_pca
pca_df['dbscan_labels'] = X_dbscan_pca
```

```python
tsne_df = pd.DataFrame(xByTSNE , columns=['x','y'])
tsne_df['kmeans_labels'] = X_kmeans_tsne
tsne_df['dbscan_labels'] = X_dbscan_tsne
```
